```python
import pandas as pd


import pandas as pd
import sqlite3

# https://www.kaggle.com/datasets/wyattowalsh/basketball
con = sqlite3.connect("data/basketball.sqlite")


df = pd.read_sql_query("""
SELECT *
FROM Player_Salary t1 
JOIN Player t2 
ON t1.namePlayer = t2.full_name 
JOIN Player_Attributes t3 
ON t2.first_name = t3.FIRST_NAME and t2.last_name = t3.LAST_NAME
""", con)

```

```python
df = df.fillna(0)
df.columns
```

```python
df['TEAM'] = df.TEAM_CITY + ' ' + df.TEAM_NAME
```
